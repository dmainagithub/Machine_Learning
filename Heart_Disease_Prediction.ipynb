### Importing dependencies


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Data collection and processing

```python
# Loading the data from csv file
heart_data = pd.read_csv('/content/heart.csv')
heart_data.head()
```

```python
# Number of rows and columns in the dataset
heart_data.shape
```

```python
# get some information about our data
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# statistical measures about heart data
heart_data.describe().T
```

```python
heart_data.describe().T
```

```python
# checking the distribution of the target variable (We should have a distribution that is almost equal, otherwise we will have class imbalance)
heart_data['target'].value_counts()
```

1 -> Defective heart

0 -> Healthy heart

### Splitting the features and the target

```python
X = heart_data.drop(columns='target', axis=1)
y = heart_data['target']
```

```python
X.head()
```

```python
print(y)
```

### Splitting the data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=2) # random_state ensures that we split our training and test data the same way - reproducibility
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Model Training

```python
model = LogisticRegression()
```

```python
# Training the logistic regression model with training data
model.fit(X_train, y_train) # it will find the relationship (pattern) between these features and the outcome of interest
```

### Model Evaluation
Accuracy score

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print("Accuracy on training data  :", training_data_accuracy)
```

```python
# Accuracy on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print("Accuracy on testing data  :", testing_data_accuracy)
```

1. Handling missing values through imputation
2. The other aspect of improving accuracy, adding more training data (with a small dataset the accuracy might be wanting)
3. The accuracy between training data and test data should not be so much, otherwise it means our model is overfitted.
